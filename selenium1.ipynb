Q1: Write a python program to scrape data for “Data Analyst” Job position in 
“Bangalore” location. You have to scrape the job-title, job-location, company_name,
experience_required. You have to scrape first 10 jobs data.


```python
#Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
```

```python
#opening the home page of naukri.com
url=("https://www.naukri.com/")
driver.get(url)
```

```python
search_button=driver.find_element_by_xpath('//span[@class="fr geoLocBtn later"]')  
search_button.click()
```

```python
#enter the input to search the keywords.
user_inp_job=input('Enter the job role')
```

```python
search_job=driver.find_element_by_id("qsb-keyword-sugg")       #locating searc_bar by id
search_job.clear()                                                 #Clearing search_bar
search_job.send_keys(user_inp_job)                               #sending user input to search bar
time.sleep(4)
```

```python
#enter the input to search the keywords.
user_inp_loc=input('Enter the loction')
```

```python
search_loc=driver.find_element_by_id("qsb-location-sugg")       #locating searc_bar by id
search_loc.clear()                                                 #Clearing search_bar
search_loc.send_keys(user_inp_loc)                               #sending user input to search bar
time.sleep(4)
```

```python
search_btn=driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
Title=[]
Location=[]
Company=[]
Experience=[]

urls=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')    #collecting urls of all the Jobs

UR=[]

for i in urls[:10]:

    UR.append(i.get_attribute('href'))                 #getting the url of first 10 Jobs

for url in UR:                                  #loop for every Jobs in the list

    driver.get(url)

    try:                                       #exception handling for no such element exception


        title=driver.find_element_by_xpath('//h1[@class="jd-header-title"]')   #location of title

        Title.append(title.text)                     #appending the title from title list
        
        company=driver.find_element_by_xpath('//div[@class="jd-header-comp-name"]')   #location of Company

        Company.append(company.text)                #appending the Company from Company list
        
        location=driver.find_element_by_xpath('//div[@class="loc"]/span')   #location of location

        Location.append(location.text)        #appending the location from location list
        
        experience=driver.find_element_by_xpath('//div[@class="exp"]/span')   #location of experience

        Experience.append(experience.text)    #appending the experience from experience list
        
    except NoSuchElementException   as e:

        Title.append("No Title")                # appending the No title if no title is there       
        
        Company.append("No Company name")       # appending the No Company name if no Company name is there
        
        Location.append("No Location")         # appending the No Location if no Location is there
        
        Experience.append("No Experience")      #appending the No Experience if no Experience is there
        

```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['Title']=Title[:10]
jobs['Company Name']=Company[:10]
jobs['Location']=Location[:10]
jobs['Experience Required']=Experience[:10]
jobs
```

Q7: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to 
go the link: 

```python
!pip install selenium

import selenium 
import pandas as pd
from selenium import webdriver
import time
user_inp=input

from selenium.common.exceptions import NoSuchElementException

driver=webdriver.Chrome("chromedriver.exe")

url=("https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace")
driver.get(url)

Rating=[]
Summary=[]
Full=[]

urls=driver.find_elements_by_xpath('//div[@class="col JOpGWq"]/a')    #collecting urls of all the Jobs

UR=[]

for i in urls[:100]:

    UR.append(i.get_attribute('href'))                 #getting the url of first 100 Reviews

for url in UR:                                  #loop for every Jobs in the list

    driver.get(url)

    try:                                       #exception handling for no such element exception


        rating=driver.find_element_by_xpath('//div[@class="_3LWZlK _1BLPMq"]')   #location of rating

        Rating.append(rating.text)                     #appending the rating from rating list
        
        summary=driver.find_element_by_xpath('//p[@class="_2-N8zT"]')   #location of Summary

        Summary.append(summary.text)                #appending the Review_summary from Summary list
        
        full=driver.find_element_by_xpath('//div[@class="t-ZTKy"]')   #location of Full

        Full.append(full.text)        #appending the full review from full list
        
        
        
    except NoSuchElementException   as e:

        Rating.append("No Rating")                # appending the No Rating if no rating is there       
        
        Summary.append("No Review Summary")       # appending the No Review Summary if no review ummary is there
        
        Full.append("No Full Review")         # appending the No Full review if no Full review is there
        


import pandas as pd
Reviews=pd.DataFrame({})
Reviews['Rating']=Rating[:100]
Reviews['Review_summary']=Summary[:100]
Reviews['Full review']=Full[:100]
Reviews
```
